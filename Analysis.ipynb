```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sunshine.csv')
df.head()
```

```python
sns.barplot(data = df, x='Row Labels',y='Sum of Sunshine Duration (hrs)')
plt.xlabel('Year')
plt.ylabel(' Sunshine Duration (hrs)')
plt.title('Sunshine Duration every year')
plt.xticks(rotation=90)
#plt.legend(False)
plt.show()
```

```python
df=pd.read_csv('ProductionData.csv')
df.rename(columns={'Row Labels': 'Year'}, inplace=True)
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Sum of all peat products'], marker='o', linestyle='-')
plt.title('Energy Production Trends by Type Over the Years')
plt.plot(df['Year'], df['Sum of all renewable energies'], marker='o', linestyle='-', label='Sum of all renewable energies')
plt.plot(df['Year'], df['Sum of all coal products'], marker='o', linestyle='-', label='Sum of all coal products')
plt.plot(df['Year'], df['Sum of all indigenous production'], marker='o', linestyle='-', label='Sum of indigenous production')
plt.plot(df['Year'], df['Grand Total'], marker='o', linestyle='-', label='Grand Total')
plt.xlabel('Year')
plt.ylabel('Energy Production')
plt.grid(True)
plt.xticks(df['Year'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
corr = df.corr()

# Generate a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
df = pd.read_csv('WindFarm.csv')
df = df.fillna(0)
df.head()
df = df.rename(columns={"Row Labels":"Year"})
plt.figure(figsize=(5,5))
df.plot(x = "Year", y=["Total"])
plt.xlabel('Year')
plt.ylabel('Production capacity per hour(GW)')
plt.title('Wind Farm production capacity')

# Display the plot
plt.show()
```

```python
df.plot(x="Year", y = "Number of windfarms installed", kind = "bar")
plt.xlabel('Year')
plt.ylabel('Number of windfarms installed')
plt.title('Number of windfarms installed each year')
```
